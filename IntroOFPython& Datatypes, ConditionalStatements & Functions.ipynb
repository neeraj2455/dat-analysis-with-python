##**History of programming language**

Ada Lovelace is often considered the world’s first computer programmer. She is best known for her work on Charles Babbage’s proposed mechanical general-purpose computer, the Analytical Engine. While Babbage designed the machine, it was Lovelace who saw its potential to do much more than just calculate numbers.

In 1843, she translated an article by the Italian mathematician Luigi Federico Menabrea, and in her notes, she added an algorithm to calculate Bernoulli numbers. This is considered the first algorithm intended to be processed by a machine, making it the first instance of computer programming.

Though there wasn’t a formal “programming language” as we understand it today, her notes on the Analytical Engine laid the groundwork for future computer programming concepts.

##**Typecasting**
####**What is Type Casting?**

Type casting (also called type conversion) is the process of converting one data type into another in Python. For example, converting an integer (int) into a floating-point number (float).

Python provides two types of type casting:

1. Implicit Type Casting – Done automatically by Python.

2. Explicit Type Casting – Done manually by the programmer using built-in functions.

####**1. Implicit Type Casting**

```python
x=10
y=5.6
print((x+y), type(x+y)) # automatically it coverts int->float
```

```python
print(type(x))
print(type(y))
```

####**2. Explicit Type Casting**
Sometimes, we need to manually change a data type using casting functions like:

* int() – Converts to an integer

* float() – Converts to a floating-point number

* str() – Converts to a string

* list() – Converts to a list

* tuple() – Converts to a tuple

* set() – Converts to a set

```python
x=101.6
y=int(x) # converts float-> int
y
```

##**Conditonal Statements**
###**What is Conditional statements?**
Conditional statements allow a program to make decisions based on certain conditions. They help execute different code depending on whether a condition is True or False.

For example, if it's raining, you take an umbrella. If it's sunny, you wear sunglasses. This is how conditions work in programming too!

Types of Conditional Statements in Python
Python provides the following conditional statements:

* if statement – Runs code only if the condition is True.

* if-else statement – Runs one block if True, another if False.

* if-elif-else statement – Checks multiple conditions.

* Nested if statement – An if inside another if.


####**1. if Statement**
The if statement runs a block of code only if the condition is True.

Ex:

```python
age = 18
if age >= 18:
    print("You are eligible to vote!")
```

####**2. if-else Statement**
The else block runs if the condition is False.

Ex:

```python
age = 16
if age >= 18:
    print("You can vote.")
else:
    print("You cannot vote yet.")
```

####**3. if-elif-else Statement (Multiple Conditions)**
The elif (short for "else if") checks multiple conditions.

Ex:

```python
marks = 85

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 60:
    print("Grade: C")
else:
    print("Grade: F")
```

##**Loops**

####**What are Loops?**
Loops are used to repeat a block of code multiple times. Instead of writing the same code again and again, we use loops to automate repetitive tasks.

**Types of Loops in Python**

Python has two main types of loops:

1. for loop – Used when you know how many times to repeat.

2. while loop – Used when you want to repeat until a condition is False.

###**1. for Loop**
A for loop is used when you know how many times you want to repeat something.

Ex: Printing Numbers from 1 to 5



```python
for i in range(1, 6):  # Loops from 1 to 5
    print(i)
```

range(1, 6) generates numbers from 1 to 5 (it stops before 6).

####**Looping Through a List**

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

###**2. while Loop**
A while loop keeps running as long as a condition is True.

Ex: Printing Numbers Until 5

```python
num = 1
while num <= 5:
    print(num)
    num += 1  # Increases num by 1 in each loop
```

The loop stops when num reaches 6 because num <= 5 becomes False.



##**Functions**
####**What is a Function?**
A function is a block of reusable code that performs a specific task. Instead of writing the same code multiple times, you can create a function and call it whenever needed.

Functions help in:

* Code reusability (Write once, use many times).
* Code organization (Keeps the code clean and readable).
* Better debugging (Fix errors in one place).

###**Creating a User-Defined Function**
To define a function in Python, use the def keyword.
